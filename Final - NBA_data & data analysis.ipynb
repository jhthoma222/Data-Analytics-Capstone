### DATA CLEANING


```python
# Creating a 'Playoffs' column for.  Did the team make the playoffs?
# * after teams name identifies whether team made the playoffs or not

nba_df["Playoffs"] = nba_df["Team"].map(lambda x: 1 if "*" in x else 0)

nba_df.head(10)
```

```python
# Remove * after I identify 'Playoff' teams

nba_df["Team"] = nba_df["Team"].str.replace("*", "")

nba_df.head()
```

```python
# Split 2021 data from 2020 data
# There is no Playoff data for 2021 and that is what we will be predicting
 
test_df = nba_df.query("Year == 2021")       # 2021 Data
model_df = nba_df.query("Year <= 2020")      # 1990 - 2020 Data
```

### Going to explore the data and try to find correlations between the different features and the teams that made the playoffs.

For the following plots, If team made the playoffs 'Blue', If they did not make the Playoffs 'Orange'

```python
#  Just doing a quick check showing the correlation between Field Goals and Field Goal percentage.

colors = {1:'blue', 0:'orange'}
plt.scatter(model_df['FG'], model_df['FGA'], c=model_df['Playoffs'].apply(lambda x: colors[x]))
plt.xlabel('Field Goals')
plt.ylabel('Field Goal Attempts')
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.8)
plt.show()

print(np.corrcoef(model_df['FG'], model_df['FGA'])* 100)
```

81% correlation between Field goals and field goal percentage.  This was an obvious correlation,  Ran this scatterplot as a test.

```python
# Plotting the field goal attempts over the years
# Wanting to see the correlation between field goal attempts and whether the team made the playoffs.

plt.scatter(model_df['Year'], model_df['FGA'], c=model_df['Playoffs'].apply(lambda x: colors[x]))
plt.xlabel('Year')
plt.ylabel('FGA')
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.8)
plt.show()

print(np.corrcoef(model_df['FGA'], model_df['Playoffs'])*100)
```

No correlation between teams that had a high field goal percentage to making the playoffs.
So just because a team shoots the ball alot, that does not help them make the playoffs.

```python
# Plotting the field goal percentage over the years
# Wanting to see the correlation between field goal percentage and whether the team made the playoffs.

plt.scatter(model_df['Year'], model_df['FG%'], c=model_df['Playoffs'].apply(lambda x: colors[x]))
plt.xlabel('Year')
plt.ylabel('FG%')
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.8)
plt.show()

print(np.corrcoef(model_df['FG%'], model_df['Playoffs']) * 100)
```

42% correlation between high field goal percentage and teams making the playoffs.  I did expect that to be higher.  
So just because a team shoots the ball with good accuracy does not guarantee them a spot in the playoffs.

```python
# Next I want to see the 3 point attempts.  From being a fan I know that the 3 point shot has increased.
# I am curious to see how the 3 point attempts correlates to making the playoffs.

plt.scatter(model_df['Year'], model_df['3PA'], c=model_df['Playoffs'].apply(lambda x: colors[x]))
plt.xlabel('Year')
plt.ylabel('3PA')
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.8)
plt.show()

print(np.corrcoef(model_df['3PA'], model_df['Playoffs'])* 100)
```

5% correleation between 3 point attempts and making the playoffs.

```python
# Lets now look at 3 point percentage.  Does making 3 point shots help you to get into the playoffs?

plt.scatter(model_df['Year'], model_df['3P%'], c=model_df['Playoffs'].apply(lambda x: colors[x]))
plt.xlabel('Year')
plt.ylabel('3P%')
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.8)
plt.show()

print(np.corrcoef(model_df['3P%'], model_df['Playoffs']) * 100)
```

24% correlation between 3 point accuracy and making the playoffs.

```python
# Lets look at how the amount of points teams make correlates to making the playoffs.

plt.scatter(model_df['Year'], model_df['PTS'], c=model_df['Playoffs'].apply(lambda x: colors[x]))
plt.xlabel('Year')
plt.ylabel('PTS')
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.8)
plt.show()

print(np.corrcoef(model_df['PTS'], model_df['Playoffs']) *100)
```

Only a 23% correlation between 3 point accuracy and a team making the playoffs

```python
# Wins will have an obvious correlation to playoffs.  But lets take a look.

plt.scatter(model_df['Year'], model_df['W'], c=model_df['Playoffs'].apply(lambda x: colors[x]))
plt.xlabel('Year')
plt.ylabel('W')
plt.grid(color='grey', linestyle='-', linewidth=0.25, alpha=0.8)
plt.show()

print(np.corrcoef(model_df['W'], model_df['Playoffs']) * 100)
```

```python
78% correlation between wins and playoffs, which is not surprising.
```

```python
# I want an overall correlation view.  Using a heat map
# Blue means positive correlation, red means negative. The stronger the color, the larger the correlation magnitude. 
# https://towardsdatascience.com/better-heatmaps-and-correlation-matrix-plots-in-python-41445d0f2bec

corr = model_df.corr()
plt.rcParams['figure.figsize'] = (20.0, 10.0)
ax = sns.heatmap(
    corr,
    vmin=-1, center = 0,
    cmap='coolwarm_r',
    square=True,
)
ax.set_xticklabels(
    ax.get_xticklabels(),
    rotation=90,
    horizontalalignment='right'
);

plt.savefig('nba_corr.png')
```

Wins obviously has the highest correlation.  Other than thant there are not strong correlations.  Just as we saw in the individual tests.  I will use all of the statistical features to test my predicitive models.
